## 2. Data Cleaning and Preprocessing

### Handling Missing Values

Our data contains some missing values, especially in the `Victim Count` column. We use the `.fillna()` method to handle these missing values. For numerical columns like `Victim Count`, we can fill the missing values with 0 to indicate no victims in those rows.



```python
#data['Victim Count'].fillna(0,inplace  =True)
```

```python
# Replace NaN values in 'Victim Count' with 0
data['Victim Count'] = data['Victim Count'].fillna(0)
```

```python
data = data.ffill(limit=12)
data.head()
```

### Dropping Irrelevant Columns

To streamline the data and focus only on relevant features, we drop columns that are either identifiers or unnecessary for the analysis. For example, columns such as `OBJECTID`, `Report Number`, and `GlobalID` provide no value for predicting crime patterns and can be safely removed.


```python
columns_drop = ['OBJECTID', 'Report Number', 'Report Date','OffID', 'GlobalID', 'CAD ID', 'Report ID','Crime Against','Was a firearm involved?', 'Offense End Date']
data_cleaned = data.drop(columns=columns_drop)
```

## 3. Feature Engineering

Feature engineering is a crucial step where we derive new features that could provide valuable predictive power to our models. We perform the following transformations:

### Temporal Feature Extraction (Month)

We start by extracting the month from the `Offense Start Date` to capture seasonal trends in crime. This temporal feature can be useful for predicting crime patterns based on the time of year.


```python
pd.options.mode.chained_assignment = None  # Turn off the warning
data_cleaned['Offense Start Date'] = pd.to_datetime(data_cleaned['Offense Start Date'], errors='coerce')
data_cleaned = data_cleaned.dropna(subset=['Offense Start Date'])
data_cleaned['Month'] = data_cleaned['Offense Start Date'].dt.month
data_cleaned = data_cleaned.drop(columns=['Offense Start Date'])
```

```python
data_cleaned
```

We now have a new column, Month, which indicates the month in which each crime occurred. This feature will help in identifying seasonal crime patterns.

## Encoding Categorical Variables

Since many of the columns in the dataset are categorical (e.g., Zone, Location Type, NIBRS Code), we need to encode these into numerical values. We use LabelEncoder for this task, which converts each unique category into a corresponding integer.

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = data_cleaned.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data_cleaned[col] = le.fit_transform(data_cleaned[col].astype(str))
    label_encoders[col] = le
```

Now, all categorical variables are encoded as integers, making them suitable for input into machine learning models.

## Target Variable Creation (Crime Classification)

To classify zones as either high-crime or low-crime, we group the data by Zone and calculate the total crime count per zone. We then set a threshold (in this case, the median) to label zones with crime counts above the median as "high crime" and those below as "low crime."

```python
crime_count_by_zone = data_cleaned.groupby('Zone').size().reset_index(name='Total Crime')
crime_threshold = crime_count_by_zone['Total Crime'].median()  # Median as threshold
crime_count_by_zone['Crime Label'] = (crime_count_by_zone['Total Crime'] > crime_threshold).astype(int)
```

```python
# Merge Labels Back into the Dataset
data_cleaned = data_cleaned.merge(crime_count_by_zone[['Zone', 'Crime Label']], on='Zone', how='left')
```

## 5. Model Building
### Model Preparation

We separate the features (X) from the target variable (y):

We split the dataset into training and test sets, with 80% of the data used for training and 20% for testing:

```python
# Prepare Features (X) and Target (y)
X = data_cleaned[['Day of the week','Day Number', 'Zone', 'Beat','Location','Location Type', 'NIBRS Code', 'NIBRS Code Name','Longitude', 'Latitude','Council District',
                  'Neighborhood','UCR Grouping','Victim Count','Month']]  # Features
y = data_cleaned['Crime Label']  # Target: Binary Classification (High/Low Crime)

#  Handle Missing Values in Features (X)
print("Missing Values in Dataset:")
print(X.isnull().sum())
```

## Neural Network Model

We implement a basic neural network using Keras, specifying the architecture, optimizer, and loss function:

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = 4700)
```

Scale the Data Because the features of a Crime take on widely different ranges of values, to help improve the training performance of our neural network model, we first apply standard scaling to our data so that after scaling, all the features are on similar scales.

Create a StandardScaler object scaler, fit scaler to x_train and use it to transform x_train.


```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
X_test = scaler.transform(X_test)
```

```python
print('x_train.shape = ', x_train.shape)
print('y_train.shape = ', y_train.shape)
print('x_val.shape = ', x_val.shape)
print('y_val.shape = ', y_val.shape)
```

```python
def  create_nn_model():
    model = Sequential([
            Input(shape=(x_train.shape[1],)),  # Input Layer
            Dense(512, activation='relu', name='Layer_1'),  # Hidden Layer 1
            Dense(512, activation='relu', name='Layer_2'),  # Hidden Layer 2
            Dense(512, activation='relu', name='Layer_3'),  # Hidden Layer 3
            Dense(512, activation='relu', name='Layer_4'),  # Hidden Layer 4
            Dropout(0.2),  # Dropout for Regularization
            Dense(1, activation='sigmoid', name='Output')  # Output Layer (Binary Classification)
            ])
   
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
create_nn_model().summary()
```

```python
# Wrap the Keras Model for scikit-learn
nn_model = KerasClassifier(
    model=create_nn_model, 
    verbose=0
)
```

## Hyperparameter Tuning for Neural Network

```python

param_grid_nn = {
    'batch_size': [32, 64],  # Training Parameters
    'epochs': [8,10,12, 14],
}
grid_search_nn = GridSearchCV(estimator=nn_model, param_grid=param_grid_nn, cv=3, scoring='accuracy', verbose=2)
grid_search_nn.fit(x_train, y_train)
```

```python
best_nn_model = grid_search_nn.best_estimator_
nn_predictions = best_nn_model.predict(X_test)
nn_accuracy = accuracy_score(y_test, nn_predictions)
```

```python
print("Best Neural Network Parameters:", grid_search_nn.best_params_)
print("Neural Network Test Accuracy:", nn_accuracy)
```

```python
m_history = create_nn_model().fit(x_train, y_train, validation_data = (x_val,y_val), epochs= 10)
```

```python
def plot_history(history):
    score = history.history['accuracy']
    val_score = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(score))

    plt.plot(epochs, score, 'b', label='Training score')
    plt.plot(epochs, val_score, 'r', label='validation score')
    plt.title('Training and validation accuracy Score')
    plt.xlabel('epoch')
    plt.ylabel('Accuracy Score')
    plt.legend()

    plt.figure()

    plt.plot(epochs, loss, 'b', label='Training loss')
    plt.plot(epochs, val_loss, 'r', label='Training and validation loss')
    plt.title('Training loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend()

    plt.show()
    
plot_history(m_history)
```

# 7. User Input for Crime Prediction

### In this step, we allow the user to input an address and predict the crime statistics for a given area.

```python
from geopy.distance import distance  # For radius-based filtering in miles
from geopy.geocoders import Nominatim  # For geocoding (address to coordinates)
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Initialize geolocator
geolocator = Nominatim(user_agent="crime_prediction_app")

# Function to Convert Address to Coordinates
def get_coordinates_from_address(address):
    """
    Converts an address into latitude and longitude using geopy.
    """
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            raise ValueError(f"Address '{address}' not found. Please enter a valid location.")
    except Exception as e:
        print(f"Error: {e}")
        return None, None

# Function to Filter Data by Radius (in miles)
def filter_data_by_radius(center_coords, radius_miles, data):
    """
    Filters the dataset to include only crimes within a given radius (in miles) from center_coords.
    """
    def is_within_radius(row):
        return distance(center_coords, (row['Latitude'], row['Longitude'])).miles <= radius_miles
    
    return data[data.apply(is_within_radius, axis=1)].copy()

# Function for User Input and Prediction
def crime_prediction_user_input(data_cleaned, model_type="xgboost", radius_miles=1):
    """
    Allows the user to input an address or coordinates, select a model, 
    and predict crime stats for a specified radius in miles.
    """
    # Get User Input
    print("\n--- Crime Prediction App ---")
    print("Enter the location details to analyze crime data.")
    address = input("Enter Address/City/Location (or press Enter to use coordinates): ").strip()

    if address:
        # Get Coordinates from Address
        latitude, longitude = get_coordinates_from_address(address)
        if latitude is None or longitude is None:
            print("Failed to retrieve coordinates. Exiting.")
            return
        print(f"Coordinates for '{address}': Latitude={latitude}, Longitude={longitude}")
    else:
        # Use User-Provided Coordinates
        try:
            latitude = float(input("Enter Latitude: ").strip())
            longitude = float(input("Enter Longitude: ").strip())
        except ValueError:
            print("Invalid input for coordinates. Exiting.")
            return
    
    # Filter Data by Radius
    print(f"\nFiltering crimes within {radius_miles} miles of the given location...")
    filtered_data = filter_data_by_radius((latitude, longitude), radius_miles, data_cleaned)
    print(f"Found {len(filtered_data)} crimes within {radius_miles} miles.")

    # If data found, perform predictions
    if not filtered_data.empty:
        # Prepare Features for Prediction
        feature_columns = X.columns  # Use the full feature set from training
        filtered_features = filtered_data.reindex(columns=feature_columns, fill_value=0)  # Fill missing features with default value
        filtered_scaled = scaler.transform(filtered_features)  # Ensure features are scaled

        # Predict Crime Labels
        if model_type.lower() == "xgboost":
            predictions = best_xgb_model.predict(filtered_scaled)
        elif model_type.lower() == "neural_network":
            predictions = (best_nn_model.predict(filtered_scaled) > 0.5).astype(int)
        else:
            print("Invalid model type selected. Exiting.")
            return

        # Add Predictions to Filtered Data
        filtered_data.loc[:, 'Predicted Crime Label'] = predictions
        high_crime_count = (predictions == 1).sum()
        low_crime_count = (predictions == 0).sum()

        # Display Stats
        print("\n--- Crime Stats ---")
        print(f"Total Crimes Analyzed: {len(filtered_data)}")
        print(f"High Crime Zones: {high_crime_count}")
        print(f"Low Crime Zones: {low_crime_count}")
        print("\nSample of Predicted Crimes:")
        print(filtered_data[['Zone', 'Longitude', 'Latitude', 'Predicted Crime Label']].head())

        # Visualization
        plt.figure(figsize=(10, 6))
        plt.scatter(filtered_data['Longitude'], filtered_data['Latitude'], c=filtered_data['Predicted Crime Label'], cmap='coolwarm', alpha=0.6)
        plt.title(f"Crime Locations within {radius_miles} Miles of {address if address else 'Entered Coordinates'}")
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.colorbar(label="Crime Label (0 = Low, 1 = High)")
        plt.grid(True)
        plt.show()
    else:
        print("No crime data found within the specified radius.")

# Example Data Preprocessing and Model Loading
scaler = StandardScaler()
scaler.fit(data_cleaned[X.columns])  # Ensure full training feature set is used for scaling

# Example Call to Crime Prediction Routine
crime_prediction_user_input(data_cleaned, model_type="neural_network", radius_miles=10)
```
